```python
import pandas as pd
import numpy as np
import cv2
import os
```

```python
df=pd.read_csv("FaceForensics++_C23/csv/FF++_Metadata_Shuffled.csv")
df.head()
```

```python
def save_frame(fullFilePath, label, filePath):
    videoCapture = cv2.VideoCapture(fullFilePath)
    fps = videoCapture.get(cv2.CAP_PROP_FPS)
    interval = int(fps * 1)

    frame_count = 0
    saved_count = 0

    video_name = os.path.splitext(os.path.basename(filePath))[0]


    output_dir = os.path.join(label, video_name)
    os.makedirs(output_dir, exist_ok=True)

    while True:
        ret, frame = videoCapture.read()
        if not ret:
            break

        if frame_count % interval == 0:
            frame_path = os.path.join(output_dir, f"frame_{saved_count}.jpg")
            print(frame_path)
            cv2.imwrite(frame_path, frame)
            saved_count += 1

        frame_count += 1

    videoCapture.release()
```

```python
videoDetails=df[['File Path','Label']]
for filePath,label in videoDetails.itertuples(index=False):
    fullFilePath=f"FaceForensics++_C23/{filePath}"
    if os.path.exists(fullFilePath):
        save_frame(fullFilePath=fullFilePath,label=label,filePath=filePath)
            
```
