```python
import pandas as pd 
import numpy as np 
import os 
import shutil
```

```python
framesFake={"Folder":[],"Frame Count":[],"File Path":[],"Label":[]}
for folder in os.listdir('dataset/full/fake'):
    countFrame=0
    for frames in os.listdir(f'dataset/full/fake/{folder}'):
        countFrame+=1
    framesFake["Folder"].append(folder)
    framesFake["Frame Count"].append(countFrame)
    framesFake['File Path'].append(f"fake/frames/{folder}")
    framesFake["Label"].append(0)

fakeFramesDF=pd.DataFrame(framesFake)
print(fakeFramesDF.shape)
fakeFramesDF.head()
```

```python
framesReal={"Folder":[],"Frame Count":[],"File Path":[],"Label":[]}
for folder in os.listdir('dataset/full/real'):
    countFrame=0
    for frames in os.listdir(f'dataset/full/real/{folder}'):
        countFrame+=1
    framesReal["Folder"].append(folder)
    framesReal["Frame Count"].append(countFrame)
    framesReal['File Path'].append(f"real/frames/{folder}")
    framesReal["Label"].append(1)

realFramesDF=pd.DataFrame(framesReal)
print(realFramesDF.shape)
realFramesDF.head()
```

```python
print(np.sum(fakeFramesDF['Frame Count']))
print(np.sum(realFramesDF['Frame Count']))
print(np.sum(fakeFramesDF['Frame Count'])+np.sum(realFramesDF['Frame Count']))
```

```python
if not os.path.exists("fakeFrame.csv"):
    fakeFramesDF.to_csv("fakeFrame.csv",index=False,encoding="utf-8")
if not os.path.exists("realFrame.csv"):
    realFramesDF.to_csv("realFrame.csv",index=False,encoding="utf-8")
```

```python
df=pd.concat([realFramesDF,fakeFramesDF],ignore_index=True)

df=df[df["Frame Count"]<32]

for path in df["File Path"]:
    if os.path.exists(path):
        shutil.rmtree(path)
```
