```python
import pandas as pd
import numpy as np
import os
import cv2
```

```python
df=pd.concat([fakeDF,realDF],ignore_index=True)
```

```python
def extractFrames(path,output_folder):
    os.makedirs(output_folder,exist_ok=True)

    cap = cv2.VideoCapture(path)

    if not cap.isOpened():
        print("Wrong Path",path)
        return
    frameCount=0
    while True:
        ret,frame=cap.read()

        if not ret:
            break

        frame_path=os.path.join(output_folder,f"frame_{frameCount}.jpg")

        cv2.imwrite(frame_path,frame)

        frameCount+=1
        
    cap.release()
```

```python
def extractSpecificFrames(path, output_folder, num_frames):
    os.makedirs(output_folder, exist_ok=True)

    cap = cv2.VideoCapture(path)


    if not cap.isOpened():
        print("Error opening video:", path)
        return
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    step = total_frames // num_frames

    frame_id = 0
    saved = 0

    while saved < num_frames:
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_id)

        ret, frame = cap.read()
        if not ret:
            break

        frame_path = os.path.join(output_folder, f"frame_{saved}.jpg")
        
        cv2.imwrite(frame_path, frame)

        saved += 1
        frame_id += step

    cap.release()
```

```python
folders=df["File Path"]
for folder in folders:
    if folder == "fake/frames/0004_fake" or folder == "real/frames/0004":
        continue
    f=folder.split("/")
    f.pop(1)
    folder="/".join(f)
    if 'fake' in folder:
        extractFrames(f"{folder}.mp4",f"dataset/full/fake/{f[1]}")
    else:
        extractFrames(f"{folder}.mp4",f"dataset/full/real/{f[1]}")
```
